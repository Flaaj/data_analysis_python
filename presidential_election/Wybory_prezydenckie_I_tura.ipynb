```python
data = pd.read_csv(temporary, delimiter=";")
```

```python
data.head()
```

```python
data_PL = data.iloc[:13800]
data_PL.columns[17:]
```

```python
col = list(data_PL.columns)
surnames = np.array(['Biedron', 'Bosak', 'Duda', 'Hołownia', 'Jakubiak', 
                     'Kosiniak-Kamysz', 'Piotrowski', 'Tanajno', 'Trzaskowski', 'Witkowski', 'Żółtek'])
col = col[:17]
col.extend(surnames)
col = np.array(col)
col
```

```python
data_PL.columns = col
```

```python
surnames = np.append(surnames, "% głosów nieważnych")
```

```python
surnames
```

```python
data_PL.replace(',', '.', inplace=False)
glosy_niewazne = data_PL[surnames].astype(float)
glosy_niewazne
```

```python
glosy_niewazne['Duda']
```

```python
for surname in surnames[:-1]:
    plt.figure(figsize=(16,8))
    plt.scatter(glosy_niewazne['% głosów nieważnych'], glosy_niewazne[surname])
    plt.xlabel("Procent głosów nieważnych")
    plt.ylabel("Procent głosów oddanych na danego kandydata")
    plt.title(surname)
    plt.savefig(surname + '.png')
```
